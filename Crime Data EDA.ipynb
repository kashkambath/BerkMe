```python
import pandas as pd
import numpy as np
import geopy
import geocoder
```

```python
crime_data = pd.read_csv('crime_data/Berkeley_PD_-_Calls_for_Service.csv')
```

```python
crime_data['EVENTDT'] = pd.to_datetime(crime_data["EVENTDT"])
```

```python
crime_data = crime_data.sort_values('EVENTDT', ascending=True)
```

```python
lat_and_lon = (
    # Remove newlines
    crime_data['Block_Location'].str.replace("\n", "\t") 
    # Extract Lat and Lon using regular expression
    .str.extract(".*\((?P<Lat>\d*\.\d*)\, (?P<Lon>-?\d*.\d*)\)",expand=True)
)
lat_and_lon
```

```python
crime_data_with_lat = pd.merge(crime_data, lat_and_lon, right_index=True, left_index=True)
```

```python
crime_data_with_lat
```

```python
crime_data_with_lat['Lat'].isnull().sum()
```

```python
crime_data_missing_info = crime_data_with_lat[crime_data_with_lat['Lat'].isnull()]
```

```python
crime_data_missing_info
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim()
```

```python
crime_data_missing_info['Geohashed_Location'] = crime_data_missing_info['BLKADDR'].apply(lambda x: geocoder.google(x))
```

```python
crime_data_missing_info['Full Location'] = crime_data_missing_info['BLKADDR'].apply(lambda x: x + ", Berkeley")
```

```python
crime_data_missing_info['Geohashed_Location'] = crime_data_missing_info['Full Location'].apply(lambda x: geocoder.google(x))
```

```python
crime_data_missing_info
```

```python
crime_data_missing_info['Lat'] = crime_data_missing_info['Geohashed_Location'].apply(lambda x: x.latlng)
crime_data_missing_info['Lon'] = crime_data_missing_info['Geohashed_Location'].apply(lambda x: x.latlng)
```

```python
crime_data_missing_info
```

```python
crime_data_missing_info_cleaned = crime_data_missing_info.dropna()
```

```python
crime_data_missing_info_cleaned
```

```python
crime_data_missing_info_cleaned['Lon'] = [x[1] for x in crime_data_missing_info_cleaned['Lon']]
```

```python
crime_data_missing_info_cleaned.drop(['Full Location', 'Geohashed_Location'], inplace=True, axis=1)
```

```python
crime_data_missing_info_cleaned
```

```python
crime_data_full_info = crime_data_with_lat.dropna()
```

```python
crime_data_full_info
```

```python
full_crime_data = pd.concat([crime_data_full_info, crime_data_missing_info_cleaned])
```

```python
full_crime_data
```

```python
crime_data_csv = full_crime_data[['OFFENSE', 'CVLEGEND', 'BLKADDR', 'Lat', 'Lon']]
```

```python
crime_data_csv.to_csv('crime_data_table.csv')
```
