```python
bus_with_loc = businesses[~businesses['B1_FULL_ADDRESS'].str.contains(r'VARIOUS')]
```

```python
bus_with_loc['B1_FULL_ADDRESS'].value_counts()
```

```python
bus_with_loc['LOCATION'] = bus_with_loc['B1_FULL_ADDRESS'].apply(lambda x: x + " , Berkeley")
```

```python
bus_with_loc.head(10)
```

```python
businesses['Business_Location'].isnull().sum()
```

```python
lat_and_lon = (
    # Remove newlines
    businesses['Business_Location'].str.replace("\n", "\t") 
    # Extract Lat and Lon using regular expression
    .str.extract(".*\((?P<Lat>\d*\.\d*)\, (?P<Lon>-?\d*.\d*)\)",expand=True)
)
lat_and_lon
```

```python
bus_data_with_loc = pd.merge(businesses, lat_and_lon, right_index=True, left_index=True)
```

```python
bus_data_missing_info = bus_data_with_loc[bus_data_with_loc['Lat'].isnull()]
```

```python
bus_data_missing_info
```

```python
bus_data_missing_info = bus_data_missing_info [~bus_data_missing_info ['B1_FULL_ADDRESS'].str.contains(r'VARIOUS')]
```

```python
bus_data_missing_info.head(10)
```

```python
bus_data_missing_info['LOCATION'] = bus_data_missing_info['B1_FULL_ADDRESS'].apply(lambda x: x + " , Berkeley")
```

```python
bus_data_missing_info['Geohashed_Location'] = bus_data_missing_info['LOCATION'].apply(lambda x: geocoder.google(x))
```

```python
bus_data_missing_info['city'] = bus_data_missing_info['Geohashed_Location'].apply(lambda x: x.city)
```

```python
bus_data_missing_info = bus_data_missing_info[bus_data_missing_info['city'] == 'Berkeley']
```

```python
bus_data_missing_info['Lat_Lng'] = bus_data_missing_info['Geohashed_Location'].apply(lambda x: x.latlng)
```

```python
bus_data_missing_info['Lat'] = bus_data_missing_info['Lat_Lng'].apply(lambda x: x[0]) 
bus_data_missing_info['Lon'] = bus_data_missing_info['Lat_Lng'].apply(lambda x: x[1]) 
```

```python
bus_data_missing_info.drop(['LOCATION', 'Geohashed_Location'], inplace=True, axis=1)
```

```python
bus_data_missing_info.drop(['Lat_Lng'], inplace=True, axis=1)
```

```python
bus_data_missing_info.head(10)
```

```python
bus_data_with_loc = bus_data_with_loc.dropna(subset=['Lat'])
```

```python
bus_data_with_loc
```

```python
cleaned_bus_data = pd.concat([bus_data_with_loc, bus_data_missing_info])
```

```python
cleaned_bus_data.drop('city', axis=1, inplace=True)
```
